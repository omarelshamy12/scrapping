```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
from word2number import w2n
```

```python
response = requests.get('https://books.toscrape.com/')
soup = BeautifulSoup(response.text, 'html.parser')

```

```python
soup.find_all('li')[2].find("ul").find_all("li")[0].find('a').get("href")
cat_link=[]
for i in soup.find_all('li')[2].find("ul").find_all("li"):
    cat_link.append("https://books.toscrape.com/"+i.find('a').get("href"))
cat_link
```

```python
with open('books.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'price', 'rating', 'category_name'])
    writer.writeheader()
    for url in cat_link:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        titles=[]
        for t in soup.find_all('h3'):
            titles.append(t.find('a').get('title'))
        
        rating=[]
        for t in soup.find_all('article'):
            rating.append(w2n.word_to_num(t.find('p').get('class')[1]))
        
        prices=[]
        for t in soup.find_all('p',{"class":"price_color"}):
            prices.append(float(t.get_text().lstrip('Â£')))
        
        cat_name=soup.find("div",{"class":"page-header action"}).find("h1").get_text()
        for i in range(len(titles)):
            name=titles[i]
            price=prices[i]
            rate=rating[i]
            writer.writerow({'name':name, 'price':price,'rating':rate,'category_name':cat_name})
```
